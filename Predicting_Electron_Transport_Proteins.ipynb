**Load Dataset**

Dataset were uploaded by lecturer in arff format. We will use panda and scipy to import and convert them to a pandas dataframe.

```python
import pandas as pd
from scipy.io import arff
```

```python
AAC_test = arff.loadarff('/content/et_tp.AAC.test.arff')
AAC_train = arff.loadarff('/content/et_tp.AAC.train.arff')
DPC_test = arff.loadarff('/content/et_tp.DPC.test.arff')
DPC_train = arff.loadarff('/content/et_tp.DPC.train.arff')
PSSM_test = arff.loadarff('/content/et_tp.PSSM.test.arff')
PSSM_train = arff.loadarff('/content/et_tp.PSSM.train.arff')
```

```python
AAC_test = pd.DataFrame(AAC_test[0])
AAC_train = pd.DataFrame(AAC_train[0])
DPC_test = pd.DataFrame(DPC_test[0])
DPC_train = pd.DataFrame(DPC_train[0])
PSSM_test = pd.DataFrame(PSSM_test[0])
PSSM_train = pd.DataFrame(PSSM_train[0])
```

```python
AAC_test.head()
```

```python
AAC_test.shape
```

```python
AAC_test_X = AAC_test.drop('class', axis=1)
AAC_test_Y = AAC_test['class']
```

```python
print(AAC_test_X.shape)
print(AAC_test_Y.shape)
```

```python
AAC_train_X = AAC_train.drop('class', axis=1)
AAC_train_Y = AAC_train['class']
DPC_test_X = DPC_test.drop('class', axis=1)
DPC_test_Y = DPC_test['class']
DPC_train_X = DPC_train.drop('class', axis=1)
DPC_train_Y = DPC_train['class']
PSSM_test_X = PSSM_test.drop('class', axis=1)
PSSM_test_Y = PSSM_test['class']
PSSM_train_X = PSSM_train.drop('class', axis=1)
PSSM_train_Y = PSSM_train['class']
```

```python
AAC_train.head()
```

The AAC (Amino acid composition) train set has 20 features and 1 variable with labeled classes of yes or no. Yes means it is an electron transport protein. No means it is non eletron transport protein. This protein plays a vital role in cellular respiration, which is a process that converts glucose into energy. Without the interaction of this protein, human body won't get enough energy for the daily lives. Similarly, other dataset derived other aspects, DPC means Dipeptide composition, PSSM means Position specific scoring matrix.

**Testing with a KNN model**

We will parse those train and test datasets to a KNN-3 classes model.

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

However, KNN requires numerical data for classification. We will convert their Y data to numerical data using LabelEncoder from sklearn.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
AAC_train_Y = label_encoder.fit_transform(AAC_train_Y)
AAC_test_Y = label_encoder.fit_transform(AAC_test_Y)
print(type(AAC_train_Y))
print(type(AAC_test_Y))
```

```python
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```

```python
DPC_train_Y = label_encoder.fit_transform(DPC_train_Y)
DPC_test_Y = label_encoder.fit_transform(DPC_test_Y)
PSSM_train_Y = label_encoder.fit_transform(PSSM_train_Y)
PSSM_test_Y = label_encoder.fit_transform(PSSM_test_Y)
```

```python
knnACC = KNeighborsClassifier(n_neighbors=3)
knnACC.fit(AAC_train_X, AAC_train_Y)
knnACC_pred = knnACC.predict(AAC_test_X)
print(accuracy_score(AAC_test_Y, knnACC_pred))
```

```python
knnACC = KNeighborsClassifier(n_neighbors=3)
cv_scores = cross_val_score(knnACC, AAC_train_X, AAC_train_Y, cv=3, scoring='accuracy')
print("Cross-validation accuracy scores:", cv_scores)
print("Average cross-validation accuracy:", cv_scores.mean())
knnACC.fit(AAC_train_X, AAC_train_Y)
knnACC_pred = knnACC.predict(AAC_test_X)
print(accuracy_score(AAC_test_Y, knnACC_pred))
```

```python
knnDPC = KNeighborsClassifier(n_neighbors=3)
cv_scores = cross_val_score(knnDPC, DPC_train_X, DPC_train_Y, cv=3, scoring='accuracy')
print("Cross-validation accuracy scores:", cv_scores)
print("Average cross-validation accuracy:", cv_scores.mean())
knnDPC.fit(DPC_train_X, DPC_train_Y)
knnDPC_pred = knnDPC.predict(DPC_test_X)
print(accuracy_score(DPC_test_Y, knnDPC_pred))
```

```python
knnPSSM = KNeighborsClassifier(n_neighbors=3)
cv_scores = cross_val_score(knnPSSM, PSSM_train_X, PSSM_train_Y, cv=3, scoring='accuracy')
print("Cross-validation accuracy scores:", cv_scores)
print("Average cross-validation accuracy:", cv_scores.mean())
knnPSSM.fit(PSSM_train_X, PSSM_train_Y)
knnPSSM_pred = knnPSSM.predict(PSSM_test_X)
print(accuracy_score(PSSM_test_Y, knnPSSM_pred))
```
